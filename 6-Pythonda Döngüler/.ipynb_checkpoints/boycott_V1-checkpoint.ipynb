```python
import pandas as pd
from telegram import Update
from telegram.ext import Updater, CommandHandler, MessageHandler, Filters, CallbackContext

# توكن البوت الخاص بك
TOKEN = "6897793371:AAE9aIQm15aDIWuzsPKz5pOY-yK1FzfN19s"

# اسم ملف النص
txt_file = 'product_list.txt'

# قائمة المنتجات المحظورة الحالية
banned_products = []

def start(update: Update, context: CallbackContext) -> None:
    update.message.reply_text('مرحبًا! أرسل اسم المنتج للتحقق.')

def check_product(update: Update, context: CallbackContext) -> None:
    user_input = update.message.text

    # تحقق مما إذا كان المستخدم أرسل اسم منتج
    if not user_input.startswith('/'):
        # تحقق من وجود اسماء منتجات محظورة في ملف نصي وأنه غير فارغ
        try:
            with open(txt_file, 'r') as file:
                banned_products = [line.strip().lower() for line in file.readlines()]
                print(f"تم قراءة المنتجات بنجاح: {banned_products}")

        except FileNotFoundError:
            print(f"تحذير: الملف {txt_file} غير موجود.")
        except Exception as e:
            print(f"حدث خطأ أثناء قراءة الملف: {e}")

    # تحقق من ما إذا كان اسم المنتج في قائمة المنتجات المحظورة
    if user_input.lower() in banned_products:
        update.message.reply_text('وفقاً للبيانات الحالية يجب تجنب استخدام هذا المنتج فهو داعم لإسرائيل وجزاك الله خيراً.')
    else:
        update.message.reply_text('وفقاً للبيانات الحالية لا يجب مقاطعة هذا المنتج، وبإمكانك استخدامه بحرية. وإذا كان لديك معلومات مغايرة لذلك بإمكانك مراسلتنا ولك جزيل الشكر.')

def main() -> None:
    updater = Updater(TOKEN, use_context=True)
    dp = updater.dispatcher

    dp.add_handler(CommandHandler("start", start))
    dp.add_handler(MessageHandler(Filters.text & ~Filters.command, check_product))

    updater.start_polling()
    updater.idle()

if __name__ == '__main__':
    main()
```
